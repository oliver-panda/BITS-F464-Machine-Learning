```python
import os #for importing handwritten images
import cv2 #for reading handwritten images
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist
```

```python
(X_train, Y_train), (X_test, Y_test)=mnist.load_data()
```

```python
X_train= tf.keras.utils.normalize(X_train, axis=1)
X_test= tf.keras.utils.normalize (X_test, axis=1)
```

```python
class ANN():
  def __init__(self, nhd, activ, n, n_iters): #n and activ will be a list that have number of neurons and the activation of each hidden layer
    self.nhd=nhd
    self.activ=activ
    self.n=n
    self.n_iters=n_iters
    

  def MakeModel(self):
    model=tf.keras.models.Sequential()
    model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
    for i in range(self.nhd):
      model.add(tf.keras.layers.Dense (self.n[i], activation=self.activ[i]))
    model.add(tf.keras.layers.Dense (10, activation='softmax'))

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])


    model.fit(X_train, Y_train, epochs=self.n_iters)
    return model
    

```

```python
m_test1=ANN(2, ['relu', 'relu'], [40, 60], 100)
model1=m_test1.MakeModel()
```
